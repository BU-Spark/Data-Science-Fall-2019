# SJC

```python
# old cases from previous group
old_cases = pd.read_csv('cases.csv')
old_cases.head()
```

```python
# Data cleaning, show cases are appealed or not

old_cases_app = old_cases[old_cases['Court Type'] == 'APPEALS COURT']
old_cases_sjc = old_cases[old_cases['Court Type'] == 'SUPREME JUDICIAL COURT']
old_cases_sjc = old_cases_sjc[['Case Name', 'Case Type', 'Has Affirm', 'Has Reverse', 'Court Type','Lower Ct Judge']]
old_cases_sjc = old_cases_sjc.rename(columns={'Lower Ct Judge': 'judge'})
old_cases_sjc.head()

old_cases_sjc["Has Affirm"]= old_cases_sjc["Has Affirm"].astype(str)
old_cases_sjc["Has Reverse"]= old_cases_sjc["Has Reverse"].astype(str)

def define_reverse(df):
    appealed = []
    for i, row in df.iterrows():
        if (row['Has Affirm']  ==   'Yes' ) & (row['Has Reverse'] ==   'nan' ):
            appealed.append('No')
        elif (row['Has Affirm']  ==   'nan' ) & (row['Has Reverse']  ==   'Yes' ):
            appealed.append('Yes')
        else:
            appealed.append('Partially')
    return appealed

old_cases_sjc['appealed'] = define_reverse(old_cases_sjc)
old_cases_sjc = old_cases_sjc.drop(['Has Reverse', 'Has Affirm', 'Court Type'], axis = 1)
old_cases_sjc.head()
```

```python
new_cases_sjc = pd.read_csv('new_sjc_data.csv')
```

```python
# Preprocessing for concating old cases dataframe and new cases dataframe
new_cases_sjc = new_cases_sjc.loc[:, ~new_cases_sjc.columns.str.contains('^Unnamed')]
new_cases_sjc = new_cases_sjc.rename(columns={'cases': 'Case Name', 'type' : 'Case Type' })
new_cases_sjc = new_cases_sjc.drop(['headnote','text','caseid'], axis = 1)
```

```python
new_cases_sjc['Case Type'] = new_cases_sjc['Case Type'].str.capitalize()
new_cases_sjc.head()
```

```python
# combine 2 dataframes together for analyzing
all_cases_sjc = pd.concat([old_cases_sjc,new_cases_sjc],axis= 0)
all_cases_sjc
```

```python
# Count cases that are affirmed, reversed, and partially reversed

total_count = all_cases_sjc.appealed.count()

reverse_count = all_cases_sjc.loc[(all_cases_sjc['appealed'] == 'Yes')].appealed.count()
affirm_count = all_cases_sjc.loc[(all_cases_sjc['appealed'] == 'No')].appealed.count()
partially_reverse_count = all_cases_sjc.loc[(all_cases_sjc['appealed'] == 'Partially')].appealed.count()
print('reverse_count = ', reverse_count, '    affirm_count = ' , affirm_count, '    partially_reverse_count = ' , partially_reverse_count )
```

## Judges

```python
new_sjc = new_cases_sjc[["Case Name", "Case Type", "appealed", "judge"]]
new_app = new_data[["cases", "type", "case status", "lower court judge"]]
new_sjc = new_sjc.rename(columns={"Case Name": "cases", "Case Type": "type","appealed" : "case status", "judge" : "lower court judge"})
```

```python
for i in range(len(new_sjc)):
    temp = new_sjc["case status"].iloc[i]
    if temp == "Partially" :
        new_sjc["case status"].iloc[i] = "partially reversed"
    elif temp == "Yes":
        new_sjc["case status"].iloc[i] = "reversed"
    elif temp == 'No':
        new_sjc["case status"].iloc[i] = "affirm"
    else:
        new_sjc["case status"].iloc[i] = "uncertian"
```

```python
all_new_data = pd.concat([new_app,new_sjc],axis= 0)
```

```python
all_new_data.head()
```

```python
for i in range(len(all_new_data)):
    temp = all_new_data["lower court judge"].iloc[i]
    if type(temp) != str or temp == "0.0" or temp == "None":
        all_new_data["lower court judge"].iloc[i] = "Not Mentioned"
    elif temp.endswith(", J."):
        all_new_data["lower court judge"].iloc[i] = all_new_data["lower court judge"].iloc[i][:-4]
```

```python
all_judge = {}
for i in range(len(all_new_data)):
    temp_name = all_new_data["lower court judge"].iloc[i]
    if temp_name not in all_judge:
        all_judge[temp_name] = [1,0,0,0]
    else:
        all_judge[temp_name][0] += 1
    temp_type = str(all_new_data["type"].iloc[i])
    temp_status = all_new_data["case status"].iloc[i]
    if temp_type.lower() == "civil" and temp_status != "affirm":
        all_judge[temp_name][1] += 1
        all_judge[temp_name][2] += 1
        all_judge[temp_name][3] += 1
    elif temp_type.lower() == "civil":
        all_judge[temp_name][2] += 1
    elif temp_status != "affirm":
        all_judge[temp_name][1] += 1
```

```python
judges = pd.DataFrame.from_dict(all_judge, orient='index', columns=['all cases', 'all reversed cases', 'civil cases', 'reversed civil cases'])
judges["reversed case rate"] =  judges["all reversed cases"]/judges["all cases"]
judges["reversed civil case rate"] = judges["reversed civil cases"]/judges["civil cases"]
```

```python
judges_top_5 = judges.sort_values(by=['all reversed cases'],ascending=False)[1:11]
```
