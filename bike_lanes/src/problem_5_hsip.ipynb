```python
import requests
```

```python
# The reason why use the API is the csv file do not contain any geo info and data info

url = "https://opendata.arcgis.com/datasets/0b1a81b2298c47739e97fa96f9c7af55_2.geojson"

# call url to store the data in r
r=requests.get(url)
# check the status
print("Stauts Code",r.status_code)

# json answer the requeset
response_dict=r.json()

#processing the data
features = response_dict['features']
# print(features[128])

# loc_128 = features[128]['geometry']['coordinates']

# location of 129 group

geo_dict = {}

for i in range(0,129):
    geo_dict[i] = []
    geo_dict[i].append(features[i]['geometry']['coordinates'][0])
    
# print(geo_dict[1][0][2])

latitudes = []
longitudes = []

for key in geo_dict.keys():
    for i in range(0,len(geo_dict[key][0])):
        lat.append(geo_dict[key][0][i][1])
        long.append(geo_dict[key][0][i][0])
    
#     print(key, " : ", val)
```

```python
#load block data


df_block = pd.read_csv("bike_block.csv")

df_block['open_dt'] = pd.to_datetime(df_block['open_dt'])
df_block['target_dt'] = pd.to_datetime(df_block['target_dt'])
df_block['closed_dt'] = pd.to_datetime(df_block['closed_dt'])

# df_block.info()

df_block.info()
```

```python
block_crash_dict = {}

for row in df_block.itertuples(index=True, name='Pandas'):
    block_crash_dict[getattr(row, "Index")]= 0
    lat =getattr(row, "latitude")
    long = getattr(row, "longitude")
    
    for key in geo_dict.keys():
        for i in range(0,len(geo_dict[key][0])):
#             bool_location = (abs(geo_dict[key][0][i][1] - lat) < 0.001) and (abs(geo_dict[key][0][i][0] - long)< 0.0001)
            bool_location = (abs(geo_dict[key][0][i][1] - lat) < 0.001) and (abs(geo_dict[key][0][i][0] - long)< 0.001)
            if(bool_location):
                block_crash_dict[getattr(row, "Index")] += 1
```
