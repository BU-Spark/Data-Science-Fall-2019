```python
ls=[]#[[]]ls # where type is nan
lsnan=[]  
import numpy as np
for i, r in new_sjc_draft.iterrows():
    if r['type'] is np.nan and r['caseid'] is not np.nan:
        ls.append([i,r['caseid']])
    if r['type'] is np.nan and r['caseid'] is np.nan:
        lsnan.append([i])
            
```

```python
# change nan to ccivil and criminal
typecriminal=[337,388,457,511,574,613]
for i in ls:
    index=i[0]
    if index in typecriminal:
        new_sjc_draft.at[index,'type']='criminal'
    else:
        new_sjc_draft.at[index,'type']='civil'
        
# type where is nan and caseid is nan to not mentioned
for i in lsnan:
    new_sjc_draft.at[i,'type']='Not Mentioned'   
```
