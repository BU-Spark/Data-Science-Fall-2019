```python
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:80% !important; }</style>"))
import pandas as pd
import numpy as np
import math
import datetime as dt
import glob
import os

# pre setting
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
## read all files in a folder

print(os.getcwd())

# find all roads with idle/s
def find_idle_count_for_roads(df):
    
    df_of_zeros_rows = df.loc[df['speed'] == 0] # return all rows with speed = 0
    list_road = np.unique(df_of_zeros_rows['road_name'],return_counts=True) # find all unique names of roads and get their counts
    output_df = pd.DataFrame(list_road[0].tolist(), columns=['road_name']) # create a new dataframe and first column is road_name 
    output_df['idle_count'] = list_road[1].tolist() # second colums is the corresponding idle_count
    
    return output_df

# output = find_idle_count_for_roads(df)

def pick_road_with_long_idle(output_df, length=5):
    df["bus_name"] = ""
    return output_df[output_df['idle_count']>length] # return all roads with idle count greater than 10, length can be manually picked

# pick_road_with_long_idle(output)

```

```python
# def process_one_day(path): 
for file in os.listdir("API_dataset/"):
    print(file)
    all_files = glob.glob("API_dataset/"+ file + "/*.csv")
    frames = []
    
    for filename in all_files:
        bus_name = filename.replace(("API_dataset/"+file+'\\'),'').replace('.csv','')
        df = pd.read_csv(filename, index_col=None, header=0)
        output = find_idle_count_for_roads(df)
        output = pick_road_with_long_idle(output)
        output["bus_name"] = bus_name.replace('API_dataset/','')
        frames.append(output)
        result = pd.concat(frames)

        if(filename==file+"\WB958.csv"): 
            break
    # print(result)
    print("done for files in", file)
    print(result)
    result.to_csv("_processed"+file+".csv")
```

```python
## return final result
all_processed_files = glob.glob("*.csv") # read all processed files

li = []
for filename in all_processed_files: # put all processed files in one dataframe
    df_temp = pd.read_csv(filename)
    li.append(df_temp)
final = pd.concat(li, axis=0, ignore_index=True)
final = final.sort_values(by=['road_name']) 

final = final.groupby('road_name')['idle_count'].sum().reset_index() # sort idle_count based on road_name and sum them up respectively
final = final[final['idle_count'] >10000]

# ax = final.plot.bar(x=final['road_name'],rot=0)
# final
final.groupby("road_name")['idle_count'].mean().plot(kind='bar')
```

```python
## find the highest idling time route
list_idlenum = []
list_bus = []
for file in os.listdir("X:/506Final/dataset/"):
    all_files = glob.glob("X:/506Final/dataset/"+ file + "/*.csv")

    for filename in all_files:
        bus_name = filename.replace(("X:/506Final/dataset/"+file+'\\'),'').replace('.csv','')
        df = pd.read_csv(filename, index_col=None, header=0)
        output = find_idle_count_for_roads(df)
        busName = bus_name.replace(file,'')
        list_idlenum.append((output['idle_count']).sum())
        list_bus.append(busName.replace(file,''))
        #print(len(list_idlenum),len(list_bus))
total_idle = pd.DataFrame(list_idlenum,columns=['idle_num'])
total_idle['bus_name'] = list_bus

total_idle = total_idle.groupby('bus_name')['idle_num'].sum().reset_index()
total_idle = total_idle.sort_values(by=['idle_num'])
total_idle.tail(10)
```
