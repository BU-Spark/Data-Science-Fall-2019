```python
import pandas as pd
```

```python
mapc_data = pd.read_csv("mapc.ma_parcels_metrofuture.csv")
mapc_data
```

```python
# 1. filter by using type of parcel which equals to FEE or TAX
mapc_fee_tax = mapc_data[(mapc_data['poly_typ'] == 'FEE') | (mapc_data['poly_typ'] == 'TAX')]
mapc_fee_tax
```

```python
# 2. filter by using standard use code
surplus_land = mapc_fee_tax[ 
        (mapc_fee_tax['luc_adj_1'] == 91.0)
    | 
        ( (mapc_fee_tax['luc_adj_1'] >= 910.0) & (mapc_fee_tax['luc_adj_1'] <= 919.0) )
    |
        (mapc_fee_tax['luc_adj_1'] == 92.0)
    | 
        ( (mapc_fee_tax['luc_adj_1'] >= 920.0) & (mapc_fee_tax['luc_adj_1'] <= 929.0) ) 
    |
        (mapc_fee_tax['luc_adj_1'] == 97.0)
    | 
        ( (mapc_fee_tax['luc_adj_1'] >= 970.0) & (mapc_fee_tax['luc_adj_1'] <= 979.0) ) 
    |
        (mapc_fee_tax['luc_adj_2'] == 91.0)
    | 
        ( (mapc_fee_tax['luc_adj_2'] >= 910.0) & (mapc_fee_tax['luc_adj_2'] <= 919.0) ) 
    | 
        (mapc_fee_tax['luc_adj_2'] == 92.0)
    | 
        ( (mapc_fee_tax['luc_adj_2'] >= 920.0) & (mapc_fee_tax['luc_adj_2'] <= 929.0) ) 
    |
        (mapc_fee_tax['luc_adj_2'] == 97.0)
    | 
        ( (mapc_fee_tax['luc_adj_2'] >= 970.0) & (mapc_fee_tax['luc_adj_2'] <= 979.0) ) 
]
surplus_land
```

```python
owner_names = pd.read_csv("owner_name.csv")['owner_name'].to_list()
owner_names
```

```python
# 3. filter by owner_name to get state owned parcels
state_owned_surplus_land = surplus_land[surplus_land['owner_name'].isin(owner_names)]
state_owned_surplus_land
```

```python
state_owned_surplus_land.to_csv("state_owned_surplus_land.csv")
```
