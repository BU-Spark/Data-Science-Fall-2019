```python
import pandas as pd
from ethnicolr import pred_census_ln

# Need to remember to install the enthnicolr package before trying to 
# run this on your own computer. 
```

```python
# import sdo mbe data
mbe = pd.read_csv("data/sdo_mbe_data/cleanedMBEData.csv", sep=",")
```

```python
mbe.head()
```

```python
# use Ethnicolr package to predict the race based on contact name
pred_census_ln(mbe, "BusinessContact_FullName")
```

```python
# report unique races
pd.unique(mbe["race"])
```

```python
# Report number of instances of each race in the dataset 
# Seems black owned businesses are vastly underpredicted or are not in the dataset at all.
# Either way this data is not useful when looking into black owned business
mbe.groupby("race").count()["Business_Name"]
```

```python
# export race prediction data
mbe.to_csv("data/sdo_mbe_data/cleanedMBE_race.csv", index=False)
```
