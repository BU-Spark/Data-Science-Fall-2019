```python
import pandas as pd 
import numpy as np
```

```python
# Import all contracts data
prime_contracts_unpro = pd.read_csv("data/usaspending_data/all_contracts_prime_awards_1.csv")
```

```python
 # Selection of attributes to be used for exploration
prime_contracts = prime_contracts_unpro[['contract_award_unique_key', 
       'obligated_amount','period_of_performance_start_date',
       'period_of_performance_current_end_date',
       'funding_agency_name', 
       'funding_sub_agency_name',              
       'recipient_name', 
       'primary_place_of_performance_county_name',
       'recipient_city_name',                                   
       'naics_description',
       'number_of_employees', 'annual_revenue',
       'multi_year_contract_code',
                                          
       'minority_owned_business',
       'black_american_owned_business',
       'hispanic_american_owned_business',
                                          
       'federally_funded_research_and_development_corp', 
       'us_state_government', 
       'sole_proprietorship', 
       'us_government_entity',
       'educational_institution', 
       'manufacturer_of_goods', 
       'limited_liability_corporation',
                                          
       'for_profit_organization',
       'nonprofit_organization', 
       'other_not_for_profit_organization']]
    
```

```python
prime_contracts.head()
```

```python
# Map 't' and 'f' values in the dataset into boolean types for easier exploration
d = {'t': True, 'f': False}
prime_contracts['minority_owned_business'] = prime_contracts['minority_owned_business'].map(d)
prime_contracts['black_american_owned_business'] = prime_contracts['black_american_owned_business'].map(d)
prime_contracts['hispanic_american_owned_business'] = prime_contracts['hispanic_american_owned_business'].map(d)
prime_contracts['for_profit_organization'] = prime_contracts['for_profit_organization'].map(d)
prime_contracts['nonprofit_organization'] = prime_contracts['nonprofit_organization'].map(d)
prime_contracts['other_not_for_profit_organization'] = prime_contracts['other_not_for_profit_organization'].map(d)

prime_contracts['federally_funded_research_and_development_corp'] = prime_contracts['federally_funded_research_and_development_corp'].map(d)
prime_contracts['us_state_government'] = prime_contracts['us_state_government'].map(d)
prime_contracts['sole_proprietorship'] = prime_contracts['sole_proprietorship'].map(d)
prime_contracts['us_government_entity'] = prime_contracts['us_government_entity'].map(d)
prime_contracts['educational_institution'] = prime_contracts['educational_institution'].map(d)
prime_contracts['manufacturer_of_goods'] = prime_contracts['manufacturer_of_goods'].map(d)
prime_contracts['limited_liability_corporation'] = prime_contracts['limited_liability_corporation'].map(d)
       
      
```

```python
# Create a start year attribute
prime_contracts['start_year'] = prime_contracts['period_of_performance_start_date'].str[-2:]
```

```python
# Clean so only start years in the specified fiscal year range are included
indices = prime_contracts[prime_contracts['start_year'].astype(int) < 10].index
indices_90s = prime_contracts[prime_contracts['start_year'].astype(int) > 20].index
prime_contracts.drop(indices, inplace=True)
prime_contracts.drop(indices_90s, inplace=True)
```

```python
# Check the correct years are being specified 
prime_contracts['start_year'].unique()
```

```python
# Create a similar end year attribute
prime_contracts['end_year'] = prime_contracts['period_of_performance_current_end_date'].str[-7:-5]
```

```python
# Select all contracts with contract amount equalling to 0 (no funds exchanged)
indices_zer0 = prime_contracts.loc[prime_contracts['obligated_amount'] == 0].index
```

```python
# drop all contracts with no funds exchanged
prime_contracts.drop(indices_zer0, inplace=True)
```

```python
# export cleaned data to csv file called contracts.csv
prime_contracts.to_csv("data/usaspending_data/contracts.csv", index=False)
```
