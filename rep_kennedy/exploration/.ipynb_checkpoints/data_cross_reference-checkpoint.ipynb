```python
import pandas as pd
import numpy as np 
```

```python
# Import Usa Spending Data
prime = pd.read_csv("data/usaspending_data/prime_awards.csv")
sub = pd.read_csv("data/usaspending_data/sub_awards.csv")
full = pd.read_csv("data/usaspending_data/full_awards.csv")
```

```python
# Import SDO Minority Data
mbe = pd.read_csv("data/sdo_mbe_data/cleanedMBE_race.csv")
```

```python
# Change all recipient names to lower case, for easier merging
prime['recipient_name_lower'] = prime['recipient_name'].str.lower()
mbe['Business_Name_lower'] = mbe['Business_Name'].str.lower()
prime['recipient_name_lower'] = prime['recipient_name'].str.lower()
```

```python
# Change recipient duns to type object
prime["recipient_duns"] = prime["recipient_duns"].astype('O')
```

```python
# Merge datasets in order to find interaction between the two datasets 
# (in order to check if there are enough interactions to use the mbe data as representation of all minority 
# owned businesses in Massachusetts)
prime_race_lower = prime.merge(mbe.dropna(subset = ["Business_Name_lower"]), how="left", left_on='recipient_name_lower', right_on='Business_Name_lower')
```
